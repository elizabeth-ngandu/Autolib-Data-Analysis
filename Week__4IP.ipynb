```
#importing pandas library
import pandas as pd
#importing numpy library
import numpy as np
```

```
#creating my dataset
url='Autolib_dataset (2) (1).csv'
df=pd.read_csv(url)
df
```

```
#Action: Checking the dataframe information
df.dtypes
```

```
#Action: Checking to see whether columns have null values
df.isnull().sum()
```

```
#Action: Dropping the columns that have null values close to the number of rows of the dataset
df.drop(["Scheduled at", "Displayed comment"], axis=1, inplace=True)
df
```

```
#Action: Cross-checking to see that all null values are removed
df.isnull().sum()
```

```
#Action: Renaming specific columns
df.rename(columns={"year": "Year", "month" : "Month", "day": "Day", "hour" : "Hour", "minute" : "Minute", "Charge Slots": "Charge slots", "Charging Status": "Charging status" }, inplace=True)
df
```

```
#Action:checking for duplicates
#Explanation: No duplicates
duplicated=df.drop_duplicates()
duplicated
```

```
#the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

new_df=df[(df['Rental status']=="operational") & (df['Bluecar counter']>0) & (df['Month']==4) & (df['City']=="Paris") & (df['Kind']=="STATION")]
new_df
print(new_df['Hour'].mode(),"\n") #most popular hour of the day for picking up
print(new_df['Hour'].value_counts().sort_values(ascending=False)) #pick the firt one
```

```
#21 is still the most popular hour for renting any car
new_df=df[(df['Rental status']=="operational") & (df['Month']==4) & (df['City']=="Paris") & (df['Kind']=="STATION")]
new_df['Hour'].mode()
```
